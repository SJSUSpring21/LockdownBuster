```python
import pandas as pd
from fbprophet import Prophet
```

```python
df=pd.read_csv("../dataset/CaliforniaVaccination.csv",usecols=[0,7])
```

```python
df.tail()
```

```python
df.describe()
```

```python
df['ds'] = pd.DatetimeIndex(df['date'])
df.drop('date', axis=1, inplace=True)
df_train = df[0:75]
```

```python
df_test = df[75:]
df_test.head()
```

```python
df.dtypes
```

```python
df_train.columns = ['y','ds']
df_train.tail()
```

```python
def 
m = Prophet(interval_width=0.95, daily_seasonality=True)
model = m.fit(df_train)
```

```python
future = m.make_future_dataframe(periods=18,freq='D')
future['cap'] = 39512345
forecast = m.predict(future)
forecast.head()
```

```python
forecast.tail()
```

```python
plot1 = m.plot(forecast)
```

```python
plt2 = m.plot_components(forecast)
```

```python
forecast[['ds','yhat']]
```

```python
df_forcasted = forecast[75:]
df_forcasted.head()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
p_error = mean_squared_error(df_test.people_fully_vaccinated,df_forcasted.yhat)
```
